```python
import cv2
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import random

%matplotlib inline
```

```python
img = cv2.imread("cameraman.png", cv2.CV_LOAD_IMAGE_GRAYSCALE)
```

```python
num_frames = 60
```

```python
video_filename = "camera_video.avi"
video = cv2.VideoWriter()
video.open(video_filename, cv2.cv.CV_FOURCC('F','M', 'P','4'), 15, img.shape, 0)
all_Hs = list()
all_accHs = list()
if video.isOpened():
    video.write(img)
    oldH = np.eye(3)
    for tt in range(1, num_frames):
        w = 0.05*(2*random.random() - 1)
        tx = 5*(2*random.random() - 1)
        ty = 5*(2*random.random() - 1)
        current_H = np.array([[1, -w, tx], [w, 1, ty], [0, 0, 1]])
        H = np.dot(current_H, oldH)

        invH = np.linalg.inv(H)
        timg = cv2.warpAffine(img, invH[0:2,:],None)
        video.write(timg)

        all_Hs.append(current_H)
        all_accHs.append(H)
        oldH = H.copy()
    video.release()
else:
      print "Cannot open", video_filename, "for writing."
```

```python
def avi2mp4(video_filename):
    mp4_video_file_name = video_filename + ".mp4"
    # Converts AVI into MP4 for loading into notebook
    # ffmpeg 
    import subprocess
    import shlex
    cmd = "ffmpeg -y -i " + video_filename + " -c:v libx264 " + mp4_video_file_name 
    retcode = subprocess.call(shlex.split(cmd))
    print cmd
    if retcode == 0:
        print "done."
    else:
        print >>subprocess.stderr
    return mp4_video_file_name
```

```python
def draw_cuboids(video_filename, locs, vals, sigma, tau, output_video_filename, edge=2, outline=0):
    '''Draws cuboids.
    video_filename: string
    locs: Nx3 numpy array with x, y, t cuboid locations
    vals: Nx1 numpy array vector with cuboid strengths
    sigma: spatial scale of cuboid
    tau: temporal scale of cuboid
    
    output_video_file_name: string
    
    edge: width in pixels of cuboid outlines
    outline: draws only cuboid outlines when 0, solid cuboids otherwise
    
    Author: Juan Carlos Niebles, 2013'''
    
    # Open input video.
    input_video = cv2.VideoCapture(video_filename)
    if not input_video.isOpened():
        print "Cannot open input video", video_filename ,"for reading."
        return None
    frame_width = np.uint32(input_video.get(cv2.cv.CV_CAP_PROP_FRAME_WIDTH))
    frame_height = np.uint32(input_video.get(cv2.cv.CV_CAP_PROP_FRAME_HEIGHT))
    
    # Open output video.
    output_video = cv2.VideoWriter()
    output_video.open(output_video_filename, cv2.cv.CV_FOURCC('F','M', 'P','4'),
                      15, (frame_height, frame_width), 1)
    if not output_video.isOpened():
        output_video.release()
        print "Cannot open output video", output_video_filename ,"for writing."
        return None
    
    # Keep track of current frame number.
    frame_number = -1
    # Remember that matrix positions from MATLAB are 1-based!
    locs = locs - 1
    
    # Cuboid color depends on strength.
    # TODO (niebles): normalize according to max and min values of input 'vals'.
    # We might also want to keep an option to set a scale manually, so that outputs from
    # multiple runs are visually comparable.
    cNorm  = mpl.colors.Normalize(vmin=0, vmax=1)
    scalarMap = plt.cm.ScalarMappable(norm=cNorm, cmap="jet")
    
    # Loop through all frames, drawing cuboids at each frame.
    while True:
        retval, frame = input_video.read()
        if (type(frame) == type(None)):
            break
        frame_number += 1
        output_frame = frame.copy()
        if not outline:
            # When drawing a solid cuboid, we operate on the HSV color space.
            gray_frame = cv2.cvtColor(frame.copy(), cv2.COLOR_BGR2GRAY)
            output_frame = np.zeros((gray_frame.shape[0], gray_frame.shape[1], 3),
                                    dtype=np.uint8)
            output_frame[:, :, 2] = gray_frame
        
        # See if we have any cuboid to draw:
        current_cubs_index = ((locs[:, 2] - tau < frame_number)*
                              (frame_number < (locs[:, 2] + tau)))
        current_locs = locs[current_cubs_index, 0:2]
        current_vals = vals[current_cubs_index]
        if current_locs.size > 0:
            # Draw cuboids.
            for row in range(current_locs.shape[0]):
                start_x = round(current_locs[row, 0] - 3*sigma)
                start_y = round(current_locs[row, 1] - 3*sigma)
                cub_width = round(6 * sigma)
                cub_color_rgba = scalarMap.to_rgba(current_vals[row])
                cub_color = (cub_color_rgba[2], cub_color_rgba[1],
                             cub_color_rgba[0])
                cub_rgb_color = np.dstack((cub_color_rgba[2], cub_color_rgba[1],
                                           cub_color_rgba[0]))
                cub_color_hsv = cv2.cvtColor(np.uint8(255*cub_rgb_color),
                                             cv2.COLOR_BGR2HSV_FULL)
                if outline:
                    # Draw cuboid outline only.
                    # Top border
                    output_frame[start_y:start_y+edge, start_x:start_x+cub_width, :] = 255*np.array(cub_color)
                    # Bottom border
                    output_frame[start_y+cub_width-edge:start_y+cub_width, start_x:start_x+cub_width, :] = 255*np.array(cub_color)
                    # Left border
                    output_frame[start_y:start_y+cub_width, start_x:start_x+edge, :] = 255*np.array(cub_color)
                    # Right border
                    output_frame[start_y:start_y+cub_width, start_x+cub_width-edge:start_x+cub_width, :] = 255*np.array(cub_color)
                else:
                    # Draw a solid cuboid
                    output_frame[start_y:start_y+cub_width, start_x:start_x+cub_width, 0] = cub_color_hsv[0, 0, 0]
                    output_frame[start_y:start_y+cub_width, start_x:start_x+cub_width, 1] = 255
        if not outline:
            output_frame = cv2.cvtColor(output_frame, cv2.COLOR_HSV2BGR_FULL)
        output_video.write(output_frame)
    print "Done."
    # Close video objects
    input_video.release()
    output_video.release()
```
