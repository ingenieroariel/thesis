## Using DenseTrajectories instead of cuboids.

This experiment will attempt to replace the use of cuboids by the use of dense trajectories, in the hopes of getting data with more information to feed it to the sparse coding step.

The methodology, paper and software for calculating the trajectories can be obtained at:

http://lear.inrialpes.fr/people/wang/dense_trajectories

## Understanding the inputs and outputs of DenseTrack

The DenseTrack utility works by being fed a video stream and outpus trajectories to the standard output (stdout). The simplest way to use it is:

```
DenseTrack subway_exit_turnstiles > trajectory.txt
```

But this file can be huge, the usual recommendation is to pipe it to gzip right away:

```
DenseTrack subway_exit_turnstiles | gzip > trajectory.gz

```

When running DenseTrack on the original video, we ended up with 4.8GB of compressed text. This uses the default values that are tuned for activity recognition and other complex tasks. Our intuition is that unusual event detection requires a lot less resolution (half a dozen trajectories per person as opposed to hundreds) which means our feature set could potentially be a lot smaller.

The inputs that we can tweak on DenseTrack are:

```
Usage: DenseTrack video_file [options]
Options:
  -h                        Display this message and exit
  -S [start frame]          The start frame to compute feature (default: S=0 frame)
  -E [end frame]            The end frame for feature computing (default: E=last frame)
  -L [trajectory length]    The length of the trajectory (default: L=15 frames)
  -W [sampling stride]      The stride for dense sampling feature points (default: W=5 pixels)
  -N [neighborhood size]    The neighborhood size for computing the descriptor (default: N=32 pixels)
  -s [spatial cells]        The number of cells in the nxy axis (default: nxy=2 cells)
  -t [temporal cells]       The number of cells in the nt axis (default: nt=3 cells)
```

We tested with different values of sampling stride (W), trajectory length (L) and neighborhood size (N) and reproduce the results below.
