## Formulating the problem

Each sample corresponds to a different cuboid, what we want to find is a representation that satisfies the following:

* There is a dictionary and a code that satisfy: $$data = dictionary * code $$
* The matrix code is sparse
* The values of the dictionary are small
* Adjacent code values (in terms of space or time are similar)

A mathematical formulation that satisfies this has been published as equation (4) in the paper [ZhaoFeiFeiXingCVPR2011] and is reproduced below:

$$
\begin{array}{c}
J(X_i, \alpha_i, D) = \min_{\alpha_{i}^{1},...,\alpha_{i}^{n_i}}{\frac{1}{2}\sum_{j}\left \| x_i^j - D\alpha_i^j \right \|_2^2 + \lambda_1 \sum_j{\left \| \alpha_i^j \right \|_1} + \lambda_2 \sum{W_{j,k}\left \| \alpha_i^j - \alpha_i^k \right \|_2^2}}
\end{array}
$$

Here, each x_i represents a data point and each alpha_i represents a cuboid and the variable j iterates over each of the samples. W_{j,k} is a weight matrix 

$$
x_i := \text{Data points, each }i\text{ represents a cuboid, it's size is (dimensions, samples)}
\\
\alpha_i := \text{Elements of the code matrix, it's size is (components, samples)}
\\
D := \text{Dictionary, it's size is dimensions, (components)}
\\
\lambda_1 := \text{Sparsity coeficient}
\\
\lambda_2 := \text{Adjacency coeficient}
\\
W_{j,k} = exp \left [ \frac{-\left \| x_j - x_k \right \|^2}{2\delta^2}-\frac{\left \| y_j - y_k \right \|^2}{2\delta^2}-\frac{-\left \| t_j - t_k \right \|^2}{2 \tau^2} \right ]
$$

> [ZhaoFeiFeiXingCVPR2011]
> Zhao, B., Fei-Fei, L., & Xing, E. P. (2011, June). Online detection of unusual events in videos via dynamic sparse coding. > In Computer Vision and Pattern Recognition (CVPR), 2011 IEEE Conference on (pp. 3313-3320). IEEE.
http://vision.stanford.edu/documents/ZhaoFeiFeiXing_CVPR2011.pdf

## Solving the problem

We will use an iterative solver, but before doing this, we need an initial guess for a dictionary. A good way to create is is using SVD and then removing or adding terms as needed.

```julia
samples, dimensions = size(data);
components = 200;
```

Swap axis, from (samples, dimensions) to (dimensions, samples)

```julia
X = zeros(dimensions, samples)
for item in 1:samples
    X[:, item] = data[item, :][:]
end
```

Init the dictionary with a sensible value, for doing that we take the outputs of the SVD factorization and trim or expand according to te desired number of components.
The output from the SVD factorization is one that satisfies:

$$
X == U * diagm(S) * V'
$$

```julia
U, S, V = svd(X);
```

Trim or zero pad the dictionary based on the desired number:

```julia
# Make it a vector instead of a matrix
S_matrix = reshape(S, 1, length(S));
D_svd = U .* S_matrix;
w, h = size(D_svd);
if components <= h
    D = D_svd[:, 1:components];
else
    D = [D_svd zeros(w, components - h)];
end;
```

#### Sparse code (sparse_code)

For solving the sparse coding part, we will use a method called Alternating Direction Method of Multipliers (ADMM) [BoydADMM2011]. In essence this let's us separate the l2-norm and l1-norm terms of the above optimization function using a combination of Dual Decomposition and Lagrange multipliers.


> [BoydADMM2011] Boyd, Stephen, et al. "Distributed optimization and statistical learning via the alternating direction method of multipliers." Foundations and Trends® in Machine Learning 3.1 (2011): 1-122.

##### General formulation for ADMM

The general formulation is:

Solve

$$
\begin{array}{c}
\text{minimize } f(x) + g(z)
\\
\text{subject to } Ax + Bz = c
\\
\text{with variables }x \in \mathbb{R}^n\text{ and }z \in \mathbb{R}^m
\\
\text{where }A \in \mathbb{R}^{p \times n}\text{, }B \in \mathbb{R}^{p \times m}\text{ and }c \in \mathbb{R}^p
\end{array}
$$

Assuming that f and g are convex.
The optimal value of the problem above will be noted by:

$$p = \inf\{{f (x) + g(z) | Ax + Bz = c}\}$$

We form the augmented Lagrangian:

$$L_ρ (x, z, y) = f (x) + g(z) + y^T (Ax + Bz − c) + (ρ/2)\|Ax + Bz − c\|_2^2$$

Then, ADMM consists of the iterations:

$$
\begin{array}{c}
x_{k+1} := arg\min_x L_\rho(x, z^k, y^k)
\\
z_{k+1} := arg\min_z L_\rho(x^{k+1}, z, y^k)
\\
y_{k+1} := y_k + \rho(Ax^{k+1} + Bz^{k+1} - c)
\end{array}
$$

Each of these subproblems is simple to solve. Because each time that we decide to solve for a particular variable, many terms from the original Lagrangian can be ignored.

##### ADMM formulation for the current problem:

```julia
function sparse_code(x, D,lambda1=1, lambda2=1, z=nothing; quiet=true, maxiter=100)
    params = Params(1, quiet, maxiter)
    code = lasso(D, x, lambda1; params=params)
end
```

```julia
function dict_update!(D, x, a)
   # Calculate the residuals
    R = -D * a
    R += x
    
    components = length(a)
    
    for k in 1:components
        D[:, k] = R * a[k, :]'

        l2_square = norm(D[:, k], 2)

        if l2_square < 1e-20
            # If the value is too small, let's replace it with random info
            # but set the coef to zero.
            D[:, k] = randn()
            a[k] = 0.0
            l2_square = sqrt(norm(D[:, k], 2))
        end

        # Normalize the dictionary
        D[:, k] /= sqrt(l2_square)

        # R <- 1.0 * U_k * V_k^T + R
        BLAS.ger!(1.0, D[:, k], [a[k]], R[:,:])
    end 
end
```

Putting it all together we have:

```julia
function dictionary_learning(X, components; quiet=true)
    # Calculate a dictionary that produces a sparse code for X with the given number of components.
    
    # Find an initial dictionary.
    U, S, V = svd(X);
    
    D_svd = U .* reshape(S, 1, length(S));

    w, h = size(D_svd);

    if components <= h
        D = D_svd[:, 1:components];
    else
        D = [D_svd zeros(w, components - h)];
    end;
    
    # Iterate over the samples
    for i in 1:length(X)
        x = X[:,i]
        # Find a sparse code with a given data 
        a = sparse_code(x, D; quiet=quiet)
        # Perform the dictionary update in place for efficiency
        dict_update!(D, x, a)
    end

    return D
end
```

```julia
@time D = dictionary_learning(X, components);
```
