We will use that 180 items array on each of the HOG values as the info to reconstruct.

```python
import numpy as np

X = training
intercept = np.mean(X, axis=0)
X -= intercept
deviation = np.std(X, axis=0)
X /= deviation
X.shape
```

```python
from admm.lasso import dict_learning

code, dictionary, errors = dict_learning(X, n_components=100,alpha=1)
```
