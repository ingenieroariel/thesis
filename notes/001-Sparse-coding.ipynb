# Measuring the effectiveness of sparse coding to detect unusual events

Sparse coding [MairalICML09][] has been used to detect unusual events in surveilance video [ZhaoFeiFeiXing_CVPR2011][], the aim of this work is to model the effectiveness of the technique to detect simulated fire in a simulated environment. The results of this model will be compared to a real life situation where smoking in a room was captured in surveillance video on an eldery care institution [ChengSpringer2012][].


[ZhaoFeiFeiXingCVPR2011]:
Zhao, B., Fei-Fei, L., & Xing, E. P. (2011, June). Online detection of unusual events in videos via dynamic sparse coding. In Computer Vision and Pattern Recognition (CVPR), 2011 IEEE Conference on (pp. 3313-3320). IEEE.
http://vision.stanford.edu/documents/ZhaoFeiFeiXing_CVPR2011.pdf

[MairalICML09]:
Mairal, J., Bach, F., Ponce, J., & Sapiro, G. (2010). Online learning for matrix factorization and sparse coding. The Journal of Machine Learning Research, 11, 19-60.
http://www.di.ens.fr/~fbach/mairal_icml09.pdf

[ChengSpringer2012]: Hong Cheng, Zicheng Liu, Yang Zhao, Guo Ye, Xinghai Sun, Real World Activity Summary for Senior Home Monitoring, Multimedia Tools and Applications, Springer, July 2012.
http://link.springer.com/content/pdf/10.1007%2Fs11042-012-1162-5.pdf

Here is an sketch of the activities to be performed:

1. Simulate pictures of primitive objects (ellipsis, squares).
2. Create a video using the object translated and rotated around the scene.
3. Generate a sparse coding dictionary using online learning trained with the video in (2).
4. Use Perlin noise to simulate an effect of fire in a part of the object and generate a few frames.
5. Use the dicionary learned in (3) to reconstruct the frames in (4)

6. Verify the representation of the unusual event has different properties (sparsity for example) than the training dataset, as published in Zhao-FeiFei.

7. Using the insight gathered in (6), create an algorithm to detect and label unusual fire events.

8. Generate new fires in different parts of the image, varying the location, size, power, duration and other parameters in the Perlin generator and study the performance of the unusual events labeler.

## b) Create a video using the object translated and rotated around the scene.

A video is nothing but a sequence of images, let's first create the canvas and then generate the sequence of images.

```python
canvas = np.zeros((300, 300), dtype=np.uint8)
```

Remember we defined '''rr''' and '''cc''' earlier, let's reuse it to create a circle moving along the diagonal.

```python
# This variable will hold the actual video
vid = []

# Let's iterate over each pixel in the diagonal
for i in range(20):
    # reset the canvas to all zeros
    canvas = np.zeros((300, 300), dtype=np.uint8)
    # and draw the circle in the middle
    canvas[rr+i*10, cc+i*10] = 1
    # now append that to the video
    vid.append(canvas)
```

Let's make sure it has 30 elements

```python
len(vid)
```

And then look at the one in the middle

```python
import matplotlib.pyplot as plt
imgplot = plt.imshow(vid[10])
```

For the fun of it, let's plot all 20 frames:

```python
frames = len(vid)
for index, img in enumerate(vid):
    figure()
    plt.imshow(img)
```

Now let's start the dictionary learning process. Starting with only the first frame, the initial step is to create a vector by using a moving window of 8x8 over the whole image

```python
from sklearn.decomposition import MiniBatchDictionaryLearning
from sklearn.feature_extraction.image import extract_patches_2d
from time import time

patch_size = (8, 8)
frame = vid[0]

data = extract_patches_2d(frame, patch_size)
```

```python
data.shape
```

Afterwards, we normalize it before using it to learn the dictionary:

```python
d = data.reshape(data.shape[0], -1)
```

```python
d = d +0.
```

```python
d -= np.mean(d, axis=0)
```

```python
d /= np.std(d, axis=0)
```

```python
d.shape
```

```python
print('Learning the dictionary...')
t0 = time()
dico = MiniBatchDictionaryLearning(n_components=200, alpha=1, n_iter=500)
V = dico.fit(d).components_
dt = time() - t0
print('done in %.2fs.' % dt)
```

```python
V
```

```python
import pylab as pl
import numpy as np

pl.figure(figsize=(8, 8))
for i, comp in enumerate(V):
    pl.subplot(20, 10, i + 1)
    pl.imshow(comp.reshape(patch_size), cmap=pl.cm.gray_r,
              interpolation='nearest')
    pl.xticks(())
    pl.yticks(())
pl.suptitle('Dictionary learned from patches\n' +
            'Train time %.1fs on %d patches' % (dt, len(data)),
            fontsize=16)
pl.subplots_adjust(0.08, 0.02, 0.92, 0.85, 0.08, 0.23)
```

```python
dico.set_params(transform_algorithm='lars', transform_n_nonzero_coefs=5)
code = dico.transform(d)
patches = np.dot(code, V)

intercept = np.mean(d, axis=0)
patches += intercept
patches = patches.reshape(len(d), *patch_size)
```

```python
from sklearn.feature_extraction.image import reconstruct_from_patches_2d

reconstructed =  reconstruct_from_patches_2d(patches, frame.shape)
```

```python
plt.imshow(reconstructed)
```

As we can see, the reconstructed image is even smoother than the original, let's reconstruct the whole video, without re-training the dictionary:

```python
out_vid = []

for index, img in enumerate(vid):
   
    patch_size = (8, 8)
    frame = vid[0]

    data = extract_patches_2d(img, patch_size)
    
    d = data.reshape(data.shape[0], -1)
    d = d +0.
    d -= np.mean(d, axis=0)
    d /= np.std(d, axis=0)
    
    dico.set_params(transform_algorithm='lars', transform_n_nonzero_coefs=5)
    code = dico.transform(d)
    patches = np.dot(code, V)

    intercept = np.mean(d, axis=0)
    patches += intercept
    patches = patches.reshape(len(d), *patch_size)
    reconstructed =  reconstruct_from_patches_2d(patches, img.shape)
    out_vid.append(reconstructed)
```

```python
for frame in out_vid:
    figure()
    plt.imshow(frame)
```

## Adding noise

```python
from noise import snoise3

octaves = 1
freq = 16.0 * octaves
size = 300
frames = 20

n = []

for z in range(frames):
    for x in range(size):
        for y in range(size):    
            n.append(snoise3(x / freq, y / freq, z / freq, octaves))
```

```python
nn = numpy.asarray(n)
the_noise = nn.reshape(frames, size, size)
```

```python
the_noise.shape
```

```python
plt.imshow(the_noise[19])
```

Now we construct an inverted triangle from the top of the circle and upwards, we increment the size of that triangle and use that to mask some perlin noise.

```python
from skimage.draw import polygon


# This variable will hold the actual video
vid = []

# Let's iterate over each pixel in the diagonal
for i in range(20):
    # reset the canvas to all zeros
    canvas = np.ones((300, 300), dtype=np.float64)
    # and draw the circle in the middle
    canvas[rr+i*10, cc+i*10] = -1
    # now append that to the video
    x = np.array([0, i+50, i+50])
    y = np.array([0, i+10, 0])
    rr2, cc2 = polygon(y, x)
    canvas[rr2+i*10, cc2+i*10] = the_noise[i, rr2+i*10, cc2+i*10]
    
    vid.append(canvas)
```

```python
frames = len(vid)
for index, img in enumerate(vid):
    figure()
    plt.imshow(img, cmap='hot')
```

```python
out_vid = []

for index, img in enumerate(vid[0:0]):
   
    patch_size = (8, 8)
    frame = vid[0]

    data = extract_patches_2d(img, patch_size)
    
    d = data.reshape(data.shape[0], -1)
    d = d +0.
    d -= np.mean(d, axis=0)
    d /= np.std(d, axis=0)
    
    dico.set_params(transform_algorithm='lars', transform_n_nonzero_coefs=5)
    code = dico.transform(d)
    patches = np.dot(code, V)

    intercept = np.mean(d, axis=0)
    patches += intercept
    patches = patches.reshape(len(d), *patch_size)
    reconstructed =  reconstruct_from_patches_2d(patches, img.shape)
    out_vid.append(reconstructed)
```
